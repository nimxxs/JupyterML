## 인공지능의 개념
* 가) 아서 사뮤엘의 정의
* 나) 미셸의 정의

### 인공지능의 범위
* 인공지능 > 머신러닝 > 딥러닝
* 인공지능 : 인공으로 만들어진 지능
* 머신러닝 : 인공지능 구현의 한 분야로 컴퓨터가 학습을  하도록 하는 알고리즘과 기술을 개발하는 분
* 딥러닝 : 머신러닝 구현의 한 분야로 인공신경망을 통해  컴퓨터가 학습을 하도록 하는 알고리즘과 기술   개발하는 분야야

### 인공지능 구현방식
* 초기에는 지식전달방식으로 구현 (top-down)
    + 개발자가 기계에게 일반적인 규칙을 알려줌
    + 심한 변화의 양상에는 잘 대처하지 못함
* 후기에는 데이터 중심으로 구현
    + 데이터를 통해 기계가 학습(규칙/패턴인식)한 뒤
    + 새로운 데이터를 통해 예측/분류 함
    + 훈련을 위한 많은 양질의 데이터가 필요

### 인공지능의 역사
* 앨런튜닝 - 다트머스 회의, AI 용어 등장 - 1차 암흑기
* 전문가 시스템 - 체스 챔피언 이김 - 인공 신경망 -
* XOR 문제 발생 - 2차 암흑기 - 심층 신경망 출현 -
* 빅데이터 - 4차산업혁명 - 바둑 프로기사 이김

### 인공지능의 미래
* 약 인공지능 vs 강 인공지능
* 인공지능의 특이점 출현 - 2030년

### 머신러닝 알고리즘 분류
* 지도학습 : 데이터와 레이블을 통해 훈련을 수행
    + 분류(범주형) / 회귀(연속형)
* 비지도학습 : 데이터만을 통해 훈련을 수행
    + 군집 / 차원축소

#### 데이터 시각화

```python
# feature들을 dataframe으로 생성
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df.info()
```

```python
### label도 dataframe에 추가
df['species'] = iris.target
df.head()
```

```python
# 숫자로 표현된 label을 원래의 의미인 텍스트로 변환 후
# speices2 컬럼으로 추가
def getSpeices(x):
    s = 'setosa'
    if x == 1: s = 'versicolor'
    elif x == 2: s = 'virginica'
    return s
```

```python
df['species2'] = df.species.apply(lambda x: getSpeices(x))
df.head()
```

```python
# 숫자로 표현된 label을 원래의 의미인 텍스트로 변환 후
# speices3 컬럼으로 추가 2
s = pd.Series(iris.target, dtype='category') # 레이블을 시리즈객체로 생성
df['species3'] = s.cat.rename_categories(iris.target_names) # 범주형으로 재변환
df.head()
```

### 보스톤 데이터 다루기

```python
# boston 데이터 불러오기
data_url = "http://lib.stat.cmu.edu/datasets/boston"
boston = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([boston.values[::2, :], boston.values[1::2, :2]])
target = boston.values[1::2, 2]

# np.hstack([boston.values[::2, :], boston.values[1::2, :2]])는 NumPy를 사용하여 데이터를 가공하는 부분입니다. 데이터를 처리하는 방법은 다음과 같습니다:
# boston.values는 Pandas DataFrame을 NumPy 배열로 변환합니다.
# [::2, :]와 [1::2, :2]를 사용하여 데이터를 가공합니다.
# [::2, :]는 짝수 행(0번 행, 2번 행, 4번 행, 등)을 선택합니다.
# [1::2, :2]는 홀수 행(1번 행, 3번 행, 5번 행, 등)에서 처음 2개의 열을 선택합니다.
# np.hstack를 사용하여 위에서 선택한 두 개의 NumPy 배열을 가로로 합칩니다.
```

```python
# 데이터프레임 생성
df2 = pd.DataFrame(data, 
                   columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
df2.head()
```

### 손글씨를 이미지화 해서 출력

```python
# 각 feature와 레이블을 튜플로 합쳐서 리스트 객체로 만듦
imgs = list(zip(digits.images, digits.target))
imgs[0]
```

```python
for ix, (img, lbl) in enumerate(imgs[:10]):
    plt.subplot(2, 5, ix+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap=plt.cm.bone, interpolation='nearest')
    plt.title(f'traning: {lbl}')

plt.tight_layout()
```
